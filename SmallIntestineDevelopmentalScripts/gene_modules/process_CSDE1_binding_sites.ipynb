```python
import csv
import random
```

```python
class Gene:
    def __init__(self, name):
        self.name = name
        self.total_matches = 0
        self.utr5_matches = 0
        self.cds_matches = 0
        self.utr3_matches = 0
        self.n_transcripts = 0
        
    def increment(self,utr5_matches,cds_matches,utr3_matches):
        self.utr5_matches += utr5_matches
        self.cds_matches += cds_matches
        self.utr3_matches += utr3_matches
        self.total_matches += utr5_matches + cds_matches + utr3_matches
        self.n_transcripts += 1
    
    def calc_mean(self):
        self.match_summary = 0
        self.mean_total_matches = self.total_matches/float(self.n_transcripts)
        if self.mean_total_matches > 0:
            self.match_summary = 1
```

```python
def read_binding_sites(infile):
    genes = {}
    with open(infile) as f:
        reader = csv.DictReader(f, delimiter = "\t")
        for row in reader:
            if row["Gene Name"] not in genes.keys():
                genes[row["Gene Name"]] = Gene(row["Gene Name"])
            genes[row["Gene Name"]].increment(int(row["5' UTR Matches"]),int(row["CDS Matches"]),int(row["3' UTR Matches"]))
        for key in genes.keys():
            genes[key].calc_mean()
        return genes
```

```python
genes = read_binding_sites('../../output/gene_modules/all_data/csde1_binding/CSDE1_binding_sites_no_header.csv')
```

```python
## function to read gene modules
def read_modules(infile):
    gene_modules = {}
    with open(infile) as f:
        for line in f:
            info = line.split(':')
            genes = [gene.strip() for gene in info[1].split(',')]
            gene_modules[info[0]] = genes
    return gene_modules
```

```python
gene_modules = read_modules("../../output/gene_modules/all_data/plots/UnbiasedGMs_topCorr_DR.genemodules.txt")
```

```python
def get_binding_sites(module_genes):
    binding_sites = 0
    total = 0
    for gene in module_genes:
        if gene in genes.keys():
            total += 1
            binding_sites += genes[gene].match_summary
    return total, binding_sites

def run_simulation(module_genes, n_sim = 1000):
    total, binding_sites = get_binding_sites(module_genes)
    sim_results = []
    for i in range(n_sim):
        random_genes = random.sample(genes.keys(),total)
        sim_results.append(get_binding_sites(random_genes)[1])
    return total, binding_sites, sim_results
```

```python
def get_csde1_binding_stats_all_modules(outfile,n_sim = 1000):
    with open(outfile,"w") as out:
        out.write("module,n_genes,n_genes_info,n_binding_sites,pval,n_binding_sites_sim")
        for key in gene_modules.keys():
            print(key)
            total, binding_sites, sim_results = run_simulation(gene_modules[key],n_sim)
            geq = [sites for sites in sim_results if sites >= binding_sites]
            pval = len(geq)/float(n_sim)
            out.write("\n" + ','.join(map(str, [key,len(gene_modules[key]), total, binding_sites, pval])) + ',' +\
                                          '|'.join([str(n) for n in sim_results]))
```

```python
random.seed(42)
get_csde1_binding_stats_all_modules('../../output/gene_modules/all_data/csde1_binding/CSDE1_GM_stats.csv',n_sim = 1000)
```
